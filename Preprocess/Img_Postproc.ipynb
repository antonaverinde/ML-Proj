```python

def prepare_T(path,poses_):
    folders = [item for item in path.iterdir() if item.is_dir()]
    filtr=[]
    for i in range(len(folders)-2):#range(0,1):#range(len(folders)-2):
        first_folder = folders[i]
        h5_files = list(first_folder.glob("*.h5"))
        h5_path=h5_files[0]
        folder_name = h5_path.stem.lower()
        folder_path = str(h5_path.parent.parent)
        with h5py.File(h5_path, "r") as f:
            data = f["film"][:]
        ############################
        #Additional filter
        if "both" in folder_name:
            filtr.append("both")
        else:
            filtr.append("one")
        is_both = "both" in folder_name
        # a1PPT,width1PPT=[0,0],[160,280]
        # a2PPT,width2PPT=[0,0],[110,280]
        # a1PCA,width1PCA=[10,0],[160,280]
        # a2PCA,width2PCA=[10,10],[110,280]
        a1PPT,width1PPT=[0,0],[320,560]
        a2PPT,width2PPT=[0,0],[220,560]
        a1PCA,width1PCA=[20,0],[320,560]
        a2PCA,width2PCA=[20,20],[220,560]
        print(poses_[i])
        #############################
        #PPT
        for j in range(2):
            if is_both:
                a,width=a2PPT,width2PPT
            else:
                a,width=a1PPT,width1PPT
        # tg=0
            tg=PCAPPTClasses_meas.TG(data,folder_name=folder_name,folder_path=folder_path,poses=poses_[i],a=a[j],width=width[j])
            tg.FFTPPT()
            tg.SavePlots_PPT()
            tg.SaveData_PPT()
        #############################
        #PCA
        for j in range(2):
            if is_both:
                a,width=a2PCA,width2PCA
            else:
                a,width=a1PCA,width1PCA
            tg=PCAPPTClasses_meas.TG(data,folder_name=folder_name,folder_path=folder_path,poses=poses_[i],a=a[j],width=width[j])
            tg.PCAMtr()
            tg.SavePlots_PCA()
            tg.SaveData_PCA()
        #############################
        #Raw
        a,width=a1PCA[1],width1PCA[1]
        tg=PCAPPTClasses_meas.TG(data,folder_name=folder_name,folder_path=folder_path,poses=poses_[i],a=a,width=width)
        tg.ICAMtr(mICA=6)
        tg.SavePlots_ICA()
        tg.SaveData_ICA()
        tg.SaveData_Raw()
        #############################
        #TSR
        # for j in range(2):
        #     if is_both:
        #         a,width=a2PPT,width2PPT
        #     else:
        #         a,width=a1PPT,width1PPT
        #     tg=PCAPPTClasses_meas.TG(data,folder_name=folder_name,folder_path=folder_path,poses=poses_taris[i],a=a[j],width=width[j])
        #     tg.TSR()
        #     tg.SavePlots_TSR()
        #     tg.SaveData_TSR()
    print(filtr)
    unique_filtr = list(dict.fromkeys(filtr))
    print(unique_filtr)
    return tg
```
